```python
# Step 2: Data Extraction
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

def parse_profile(profile_text):
    doc = nlp(profile_text)

    # Extract education, experience, and skills
    education = []
    experience = []
    skills = []

    for ent in doc.ents:
        if ent.label_ == "ORG" and "University" in ent.text:
            education.append(ent.text)
        elif ent.label_ == "DATE":
            experience.append(ent.text)
        elif ent.label_ == "SKILL":
            skills.append(ent.text)

    return education, experience, skills

def calculate_years_of_experience(experience_dates):
    # Simplified calculation (just a placeholder)
    years = []
    for date in experience_dates:
        if "-" in date:
            start_year, end_year = map(int, date.split(" - "))
            years.append(end_year - start_year)
        else:
            years.append(0)
    return sum(years)

# Example profile text
profile_text = """
John Doe
123 Main Street
Anytown, CA 12345
(123) 456-7890
john.doe@email.com

Summary:
Seasoned software engineer with 10+ years of experience in developing and maintaining web applications. Proven ability to work independently and as part of a team, and to deliver high-quality results on time and within budget.

Education:
- Master of Science in Computer Science, Stanford University, 2012
- Bachelor of Science in Computer Science, University of California, Berkeley, 2010

Experience:
- Software Engineer, Google, 2015 - Present
  - Developed and maintained web applications for Google Search, including the search results page and the Google Maps interface.
  - Worked closely with product managers, designers, and other engineers to gather requirements, design solutions, and implement new features.
  - Led a team of 5 engineers in the development of a new mobile app for Google Maps.
- Software Engineer, Microsoft, 2013 - 2015
  - Developed and maintained web applications for Microsoft Office, including Word, Excel, and PowerPoint.
  - Worked closely with product managers, designers, and other engineers to gather requirements, design solutions, and implement new features.
  - Contributed to the development of a new cloud-based version of Microsoft Office.

Skills:
- Python
- Java
- JavaScript
- HTML/CSS
- SQL
- NoSQL
- Agile development
- Scrum
- Kanban
- Test-driven development
- Version control (Git)
"""

education, experience, skills = parse_profile(profile_text)
years_of_experience = calculate_years_of_experience(experience)

print("Education:", education)
print("Experience:", experience)
print("Skills:", skills)
print("Years of Experience:", years_of_experience)

# Step 3: Question Generation
import random

def generate_questions(skills, experience_level):
    questions = []
    for skill in skills:
        if experience_level < 5:
            questions.append(f"Basic question about {skill}?")
        else:
            questions.append(f"Advanced question about {skill}?")
    return questions

# Example
experience_level = years_of_experience
questions = generate_questions(skills, experience_level)
print("Generated Questions:", questions)

# Step 4: Assessment Scoring
def score_answers(answers, correct_answers):
    score = 0
    for answer, correct_answer in zip(answers, correct_answers):
        if answer == correct_answer:
            score += 1
    return (score / len(correct_answers)) * 100

# Example answers (user input simulation)
user_answers = ["Basic answer", "Advanced answer"]
correct_answers = ["Basic answer", "Advanced answer"]

final_score = score_answers(user_answers, correct_answers)
print("Final Score:", final_score)

# Step 5: Output Report
def generate_report(education, experience, skills, questions, answers, score):
    report = {
        "Education": education,
        "Experience": experience,
        "Skills": skills,
        "Questions": questions,
        "Answers": answers,
        "Score": score
    }
    return report

def print_report(report):
    print("**Applicant Report**")
    print("\n**Education:**")
    for degree in report["Education"]:
        print(f"- {degree}")

    print("\n**Experience:**")
    for job in report["Experience"]:
        print(f"- {job}")

    print(f"\n**Skills:** {', '.join(report['Skills'])}")

    print("\n**Questions and Answers:**")
    for question, answer in zip(report["Questions"], report["Answers"]):
        print(f"Q: {question}")
        print(f"A: {answer}")

    print(f"\n**Final Score:** {report['Score']}%")

# Generate the report
report = generate_report(education, experience, skills, questions, user_answers, final_score)

# Print the report
print_report(report)
```
